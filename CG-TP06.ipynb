```
def keyboard(key, x, y):
    global eye_x
    step = 0.2

    if key == b'a':
        eye_x -= step
    elif key == b'd':
        eye_x += step
    elif key in (b'\x1b', b'q'):  # ESC or q
        try:
            glutLeaveMainLoop()
        except Exception:
            sys.exit(0)

    glutPostRedisplay()
```
